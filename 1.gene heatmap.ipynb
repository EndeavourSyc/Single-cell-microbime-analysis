```python
import pandas as pd
from functools import reduce
import scanpy as sc
adata_ct = sc.read_h5ad("./aggr_ct.h5ad")
import matplotlib.pyplot as plt
```

```python
marker_dict = {"B":["BANK1", "MS4A1"], "T": ["CD3E", "TRAC"], "Plasma": ["IGLC1", "MZB1"]
              , "Macrophage": ["LYZ", "CD68"], "Monocyte": ["TPSB2", "CPA3"]
               , "Fibroblast": ["COL1A2", "A2M"], "Endothelial": ["CAPS", "DRC3"]
              , "Epithelial": ["NAPSA", "MUC1"]}
```

```python
adata_ct.uns["dendrogram_CellType"]['categories_ordered'] = ['Macrophage', 'Fibroblast', 'Endothelial', 'Epithelial', 'Plasma',
       'Monocyte', 'B', 'T']
```

```python
fig,ax=plt.subplots(figsize=(8,4),constrained_layout=True, dpi = 500)
sc.pl.dotplot(adata_ct, marker_dict, groupby = "CellType",var_group_labels=[""], standard_scale='var', var_group_rotation=15
              ,dendrogram=True,ax = ax,save="CellType_heatmap_dot")
# ax.savefig("./Result/CellType_heatmap_dot.pdf", dpi=300)
```

```python
fig,ax=plt.subplots(figsize=(8,4),constrained_layout=True, dpi = 500)
sc.pl.matrixplot(adata_ct, marker_dict, groupby = "CellType", dendrogram=True, use_raw = False,cmap="PuBu", var_group_rotation=15, swap_axes=False
                 , standard_scale='var',ax=ax,save="CellType_heatmap_matrix")
```

```python

ax = sc.pl.heatmap(adata_ct, marker_dict, groupby = "CellType", standard_scale='var',dendrogram=True)
```
