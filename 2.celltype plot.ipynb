```python
import pandas as pd
from functools import reduce
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt
```

```python
adata = sc.read_h5ad("./aggr_ct.h5ad")
```

```python
celltype = list(np.unique(adata.obs['CellType']))

mem = {'Non-Tumor':{},'Primary':{},'Brain':{}}
cell_type = ['B','T','Macrophage','Monocyte','Plasma','Endothelial','Epithelial','Fibroblast']

for k in mem.keys():
    for ct in celltype:
        mem[k][ct] = 0.0
        
for ct in celltype:
    for k in mem.keys():
        tmp_total = adata[adata.obs['Tissue'].isin([k])].shape[0]
        tmp_n =  adata[adata.obs['Tissue'].isin([k])&adata.obs['CellType'].isin([ct])].shape[0]
        mem[k][ct] = round(tmp_n/tmp_total * 100, 5)
```

```python
df = pd.DataFrame(columns=cell_type)
```

```python
for i,k in enumerate(mem.keys()):
    tmp_df = pd.DataFrame(mem[k].values(),index=mem[k].keys(), columns=[k]).T
    df = pd.concat([df,tmp_df])
```

```python
for i in range(1,df.shape[1]):
    df.iloc[:,i] += df.iloc[:,i-1]
df = round(df,1)
```
